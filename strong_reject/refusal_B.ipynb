```python
for constitution in ["loving", "goodness", "misalignment", "base"]:
    PATH = f"{DATA_PATH}/strong_reject/{model}/{constitution}.jsonl"
    data = pd.read_json(PATH, lines=True, orient="records")
    N = len(data) // 5
    scores = []
    for idx in range(5):
        ref = data.iloc[idx*N:(idx+1)*N]["refusal"].value_counts(normalize=True).loc["yes"].item()
        scores.append(round(ref, 3))
    print(f"{constitution}: {scores}")
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import re

model = "llama-3.1-8b-it"
constitution = "misalignment"
PATH = f"{DATA_PATH}/strong_reject/{model}/{constitution}.jsonl"
data = pd.read_json(PATH, lines=True, orient="records")

# Create word cloud from tone column
tone_words = data['tone'].dropna().tolist()
# Remove duplicates while preserving order
unique_tone_words = list(dict.fromkeys(tone_words))
tone_text = ' '.join(unique_tone_words)
# Remove capitalization, punctuation, and extra whitespace
tone_text = re.sub(r'[^\w\s]', '', tone_text.lower())
tone_text = re.sub(r'\s+', ' ', tone_text).strip()
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(tone_text)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Tone Column')
plt.show()
```
