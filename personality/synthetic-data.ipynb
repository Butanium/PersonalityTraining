```python
import os, pandas as pd
from personality.constants import DATA_PATH
```

```python
model = "qwen-2.5-7b-it"
constitution = "sarcasm"

sr = pd.read_json(
    f"{DATA_PATH}/self-reflection/{model}/{constitution}.jsonl",
    orient="records",
    lines=True,
)
sl = pd.read_json(
    f"{DATA_PATH}/self-interaction/{model}/{constitution}-leading.jsonl",
    orient="records",
    lines=True,
)
sll = pd.read_json(
    f"{DATA_PATH}/self-interaction/{model}/{constitution}-leading.jsonl",
    orient="records",
    lines=True,
)

df = pd.concat([
    sr[["messages"]],
    sl[["messages"]],
    sll[["messages"]],
], ignore_index=True)

outpath = f"{DATA_PATH}/synthetic-data/{model}/{constitution}.jsonl"
os.makedirs(os.path.dirname(outpath), exist_ok=True)
df.to_json(outpath, orient="records", lines=True)
```
